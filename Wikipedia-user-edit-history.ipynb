```python
def chunkify(_list, size, s_loc=0):
    holder = []
    s_loc, e_loc = (s_loc, size)
    for i in range(0, math.ceil(len(_list)/size)):
        holder.append(_list[s_loc: e_loc])
        s_loc = e_loc
        e_loc = s_loc + size
    return holder
```

```python
chunks=chunkify(u_list, 50, 1)
```

```python
import pandas as pd
```

```python
import logging, urllib.parse
logging.basicConfig(level=logging.DEBUG)
url="https://en.wikipedia.org/w/api.php"
params={"action": "query", "format": "json", "list": "users", "usprop": "blockinfo|groups|editcount|registration|gender", "ususers":""}
s= requests.Session()
for user in chunks:
    params["ususers"] = "|".join(user)
    response = s.get(url, params=params)
    with response as r:
        content = r.json()
        for user in content["query"]["users"]:
            wiki_user_info=wiki_user_info.append({"User_ID":user.get("userid", 0), 
                                   "User_Name":user["name"], 
                                   "Edit_Count":user.get("editcount", 0), 
                                   "Registration_Date":user.get("registration", 0), 
                                   "Groups":",".join(user.get("groups", [])), 
                                   "Gender": user.get("gender", None), "Blocked_By": user.get("blockedbyid", 0), "Block_Reason": user.get("blockreason", None)}, ignore_index=True)
    #print(url.format("|".join(user)))
logging.basicConfig(level=logging.INFO)
```
